```python
def f(x):
    return 0.2 * x**3 + 1.4 * x**2 - 8 * x - 23

def bisection(a, b, tol):
    if f(a) * f(b) >= 0:
        print("Bisection method fails.")
        return None
    c = a
    while (b - a) >= tol:
        c = (a + b) / 2
        if f(c) == 0.0:
            break
        if f(c) * f(a) < 0:
            b = c
        else:
            a = c
    return c

# Contoh penggunaan
a = -10  # batas bawah
b = 10   # batas atas
tolerance = 0.0001
root = bisection(a, b, tolerance)
print(f"Akar dari persamaan tersebut adalah: {root}")
```

```python
def false_position(a, b, tol):
    if f(a) * f(b) >= 0:
        print("False position method fails.")
        return None
    c = a
    while abs(f(c)) > tol:
        c = b - (f(b) * (b - a)) / (f(b) - f(a))
        if f(c) == 0:
            break
        elif f(c) * f(a) < 0:
            b = c
        else:
            a = c
    return c

# Contoh penggunaan
a = -10  # batas bawah
b = 10   # batas atas
tolerance = 0.0001
root_fp = false_position(a, b, tolerance)
print(f"Akar dari persamaan dengan metode false position adalah: {root_fp}")
```

```python
# Fungsi dari persamaan yang diberikan
def f(x):
    return 0.2 * x**3 + 1.4 * x**2 - 8 * x - 23

# False Position Method
def false_position(xl, xu, err_m):
    iterasi = 0
    while True:
        xr = xu - (f(xu) * (xu - xl)) / (f(xu) - f(xl))
        err = abs(f(xr))

        if err < err_m:
            break

        if f(xr) * f(xl) < 0:
            xu = xr
        else:
            xl = xr
        iterasi += 1
    return xr, iterasi

# Bisection Method (Metode Numerik Dasar)
def bisection(xl, xu, err_m):
    iterasi = 0
    while True:
        xr = (xu + xl) / 2
        err = abs(f(xr))

        if err < err_m:
            break

        if f(xr) * f(xl) < 0:
            xu = xr
        else:
            xl = xr
        iterasi += 1
    return xr, iterasi

# Parameter awal yang sama
xl = -5  # Contoh batas bawah
xu = 5   # Contoh batas atas
err_m = 0.0001  # Kondisi error maksimum

# Menjalankan kedua metode
akar_false_pos, iter_false_pos = false_position(xl, xu, err_m)
akar_bisection, iter_bisection = bisection(xl, xu, err_m)

print(f"Akar menggunakan False Position: {akar_false_pos}, Jumlah iterasi: {iter_false_pos}")
print(f"Akar menggunakan Bisection: {akar_bisection}, Jumlah iterasi: {iter_bisection}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Inisialisasi variabel
x_awal = -2.0
x_akhir = -1.0
dx    = 0.00000001
err_m = 0.0000001

x = x_awal
iterasi = 0
y = 2*x**3 - 3*x**2 + 5*x + 3

# Loop utama untuk mencari akar
while (x < x_akhir and y**2 > err_m**2):
    iterasi += 1
    x += dx
    y = 2*x**3 - 3*x**2 + 5*x + 3
    print(f"Iterasi ke-{i}")
    print(f"x = {x}")
    print(f"y = {y}")

# Mencetak variabel
print("\nVARIABLE")
print(f"x_awal = {x_awal}")
print(f"x_akhir = {x_akhir}")
print(f"dx = {dx}")
print(f"Error margin untuk y = {err_m}")
print("\nHASIL AKHIR")

# Memeriksa apakah pencarian akar berhasil
if (y**2 <= err_m**2):
    print("Pencarian akar berhasil.")
else:
    print("Pencarian akar gagal, silakan perkecil dx atau perbesar error margin.")

print(f"Banyaknya iterasi: {i}")
print(f"x = {x}")
print(f"y = {y}")
print(" ")

# Visualisasi fungsi
x = np.linspace(-2, 3, 10000)
plt.figure(figsize=(6, 6.5))
y = 2*x**3 - 3*x**2 + 5*x + 3

# Plot fungsi
plt.plot(x, y, '-g', label='y = 2*x^3 - 3*x^2 + 5*x + 3')

# Garis horizontal pada y=0
plt.axhline(0, color='black', linewidth=1)

# Plot titik potong
plt.scatter(x, y, color='red', zorder=5, label=f'Titik Potong (x = {x:.5f}, y = {y:.5f})')

# Tambahkan judul dan label
plt.title('Grafik Fungsi y = 2x^3 - 3x^2 + 5x + 3 dengan Titik Potong')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Tampilkan grafik
plt.show()
```
